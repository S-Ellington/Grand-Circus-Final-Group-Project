Data Cleaning for DRM data

```python
df_drm = pd.read_excel('/content/2022 TS2.1 Service Data and Operating Expenses Time Series by Mode_0.xlsx', sheet_name='DRM', dtype={'NTD ID': str})
```

```python
df_drm['NTD ID'] = df_drm['NTD ID'].str.zfill(5)
```

```python
df_drm.head(30)
```

```python
df_drm.columns = df_drm.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
df_drm = df_drm[(df_drm['Mode'] == 'LR')]
```

```python
df_drm = df_drm[(df_drm['Mode Status'] == 'Operating')]
```

```python
df_drm.columns
```

```python
df_drm = df_drm.drop(columns=['Last Report Year','Legacy NTD ID','Agency Status','Reporter Type','Reporting Module',\
                     'Census Year','Primary UZA Name','UACE Code','2022 Status','Mode','Mode Status'])
```

```python
df_drm = df_drm.drop(808)
```

```python
df_drm
```

```python
df_melted = pd.melt(df_drm, id_vars=['NTD ID', 'Agency Name','City','State','UZA Area SQ Miles','UZA Population','Service'], var_name='Year', value_name='DRM')
```

```python
df_melted
```

```python
df_melted['length_of_system'] = df_melted['DRM'] / 2
```

```python
df_melted.head(23)
```

```python
df_melted.to_csv('df_drm_v2.csv', index=False)
```
